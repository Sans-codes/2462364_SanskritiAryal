Task 3: Outlier Detection

```python
#Outlier Detection in HDI and GNI per Capita
outlier_dict = {}

for col in ['hdi', 'gross_inc_percap']:
    outlier_dict[col] = {}
    for year in hdi_south_asia_df['year'].unique():
        data = hdi_south_asia_df[hdi_south_asia_df['year'] == year][col]
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = data[(data < lower) | (data > upper)]
        outlier_dict[col][year] = outliers.index.tolist()

outlier_dict
```

Using the 1.5 × IQR rule, no statistically significant outliers were detected
in either HDI or GNI per Capita for South Asian countries across the years
2020, 2021, and 2022.

This indicates that development indicators within South Asia are relatively
clustered, with no country exhibiting extreme deviation from the regional trend.
The absence of outliers reflects the robustness of the IQR method, particularly
for small datasets.


```python
plt.figure(figsize=(8,6))
for year in hdi_south_asia_df['year'].unique():
    df_year = hdi_south_asia_df[hdi_south_asia_df['year'] == year]
    plt.scatter(df_year['gross_inc_percap'], df_year['hdi'], label=f"{year}", alpha=0.7)

# Highlight outliers in red
for col in ['hdi', 'gross_inc_percap']:
    for year in outlier_dict[col]:
        idx_list = outlier_dict[col][year]
        if len(idx_list) > 0:
            plt.scatter(
                hdi_south_asia_df.loc[idx_list, 'gross_inc_percap'],
                hdi_south_asia_df.loc[idx_list, 'hdi'],
                color='red',
                edgecolor='black',
                s=80,
                label='Outlier'
            )

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('HDI vs GNI per Capita with Outliers Highlighted (South Asia)')
plt.legend()
plt.grid(True)
plt.savefig("HDI_GNI_Outliers.png", dpi=300, bbox_inches='tight')
plt.show()
```

Although the scatter plot includes logic to highlight outliers, no points appear in a separate color because no observations satisfied the IQR-based outlier criteria.

```python
# Define Middle East countries
middle_east_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait","Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia","Syria", "United Arab Emirates", "Yemen"]

# Define South Asia countries
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filter data for years 2020–2022
hdi_2020_2022 = hdi_df[hdi_df['year'].between(2020, 2022)]

# Create regional subsets
hdi_south_asia_2020_2022 = hdi_2020_2022[hdi_2020_2022['country'].isin(south_asia_countries)]

hdi_middle_east_2020_2022 = hdi_2020_2022[hdi_2020_2022['country'].isin(middle_east_countries)]

# Save datasets
hdi_south_asia_2020_2022.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
hdi_middle_east_2020_2022.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)
```

This task creates two regional subsets (South Asia and Middle East) from the
cleaned dataset used in Problem 1B, covering the years 2020–2022.
The subsets are saved for use in subsequent analysis.

```python
# Aggregate HDI by country (mean across years)
sa_hdi_avg = hdi_south_asia_2020_2022.groupby('country')['hdi'].mean()
me_hdi_avg = hdi_middle_east_2020_2022.groupby('country')['hdi'].mean()

# Identify top and bottom performers
sa_top3 = sa_hdi_avg.sort_values(ascending=False).head(3)
sa_bottom3 = sa_hdi_avg.sort_values().head(3)

me_top3 = me_hdi_avg.sort_values(ascending=False).head(3)
me_bottom3 = me_hdi_avg.sort_values().head(3)

# output result
print("South Asia – Top 3 HDI Countries")
print(sa_top3, "\n")

print("South Asia – Bottom 3 HDI Countries")
print(sa_bottom3, "\n")

print("Middle East – Top 3 HDI Countries")
print(me_top3, "\n")

print("Middle East – Bottom 3 HDI Countries")
print(me_bottom3)
```

Task 6: Correlation Analysis

```python
# Selected metrics
metrics = ['gender_development', 'life_expectancy']

# Dictionary to store correlations
correlations = {}

# Compute correlations for each region
for region_name, df in {
    "South Asia": hdi_south_asia_2020_2022,
    "Middle East": hdi_middle_east_2020_2022
}.items():
    correlations[region_name] = {}
    for metric in metrics:
        corr_value = df['hdi'].corr(df[metric])
        correlations[region_name][metric] = corr_value
        print(f"{region_name} – HDI vs {metric}: {corr_value:.3f}")
```

Task 7: Outlier Detection

```python
#Detect outliers
def detect_outliers(series):
    """
    Detects outliers in a numeric pandas Series using the 1.5 × IQR rule.
    Returns the indices of the outlier values.
    """
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series[(series < lower_bound) | (series > upper_bound)]

# South Asia
sa_hdi_outliers = detect_outliers(sa_hdi_avg)
sa_gni_outliers = detect_outliers(hdi_south_asia_2020_2022['gross_inc_percap'])

# Middle East
me_hdi_outliers = detect_outliers(me_hdi_avg)
me_gni_outliers = detect_outliers(hdi_middle_east_2020_2022['gross_inc_percap'])

print("South Asia HDI Outliers:\n", sa_hdi_outliers)
print("South Asia GNI Outliers:\n", sa_gni_outliers)
print("Middle East HDI Outliers:\n", me_hdi_outliers)
print("Middle East GNI Outliers:\n", me_gni_outliers)
```

In this task, we detect outliers in HDI and GNI per Capitafor each region
(South Asia and Middle East) using the 1.5 × IQR rule.

Using the 1.5 × IQR rule:

- **South Asia:** No HDI or GNI per Capita outliers were detected.
  This indicates that the development indicators in South Asia are
  relatively clustered, with no country significantly deviating
  from the regional trend.

- **Middle East:** Yemen is an HDI outlier, with a value of 0.426,
  significantly lower than other countries in the region. No GNI per
  Capita outliers were detected. This highlights extreme disparities
  in human development within the Middle East, where economic wealth
  may not fully reflect overall development outcomes.

```python
def plot_outliers(df, region_name, hdi_outliers, gni_col='gross_inc_percap'):
    """
    Creates a scatter plot of GNI per Capita vs HDI for a region,
    highlighting HDI outliers in red.
    """
    plt.figure(figsize=(8,6))

    # Plot all countries in blue
    plt.scatter(df[gni_col], df['hdi'], color='blue', alpha=0.7, label='Countries')

    # Highlight HDI outliers in red
    if not hdi_outliers.empty:
        # Create boolean mask to match DataFrame rows
        outlier_mask = df['country'].isin(hdi_outliers.index)
        plt.scatter(
            df.loc[outlier_mask, gni_col],
            df.loc[outlier_mask, 'hdi'],
            color='red',
            edgecolor='black',
            s=80,
            label='HDI Outliers'
        )

    plt.xlabel('GNI per Capita')
    plt.ylabel('HDI')
    plt.title(f'HDI vs GNI per Capita with Outliers ({region_name})')
    plt.legend()
    plt.grid(True)
    # Save figure for report
    plt.savefig(f"HDI_GNI_Outliers_{region_name.replace(' ','')}.png", dpi=300, bbox_inches='tight')
    plt.show()

# South Asia
plot_outliers(hdi_south_asia_2020_2022, 'South Asia', sa_hdi_outliers)

# Middle East
plot_outliers(hdi_middle_east_2020_2022, 'Middle East', me_hdi_outliers)
```

Outliers in the Middle East primarily reflect extreme income levels,
whereas South Asia shows minimal outlier behavior due to more clustered
development indicators.

**South Asia:**  
- No HDI or GNI per Capita outliers were detected.  
- Development indicators are relatively clustered across countries.  
- Scatter plot shows all points in blue, confirming uniformity.

**Middle East:**  
- Yemen is an HDI outlier with a value of 0.426.  
- This country deviates significantly from the regional trend in human development.  
- No GNI per Capita outliers were detected; most countries, including wealthy Gulf states,
  have GNI within the expected IQR range.  
- Scatter plot shows Yemen as a red point, emphasizing its low HDI relative to its region.

**Significance:**  
- Outliers highlight countries with disproportionate development outcomes.  
- Yemen’s low HDI indicates severe human development challenges relative to regional averages.  
- South Asia’s lack of outliers indicates more uniform development levels, while the Middle East shows greater disparities.
