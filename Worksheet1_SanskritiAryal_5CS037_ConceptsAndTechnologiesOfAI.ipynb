3. Dot Product   
• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.  

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

4. Matrix Multiplication

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
import time
import numpy as np

# Python lists with 1 million numbers
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))
```

```python
# Element-wise Python list addition
start = time.time()
py_add = [a + b for a, b in zip(list1, list2)]
end = time.time()
print("Python addition time:", end - start)
```

```python
# Element-wise Numpy addition
arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)

start = time.time()
np_add = arr1 + arr2
end = time.time()
print("Numpy addition time:", end - start)
```

```python
# Element-wise Python multiplication
start = time.time()
py_mul = [a * b for a, b in zip(list1, list2)]
end = time.time()
print("Python multiplication time:", end - start)
```

```python
# Element-wise Numpy multiplication
start = time.time()
np_mul = arr1 * arr2
end = time.time()
print("Numpy multiplication time:", end - start)
```

```python
# Python dot product (manual)
start = time.time()
py_dot = sum(a * b for a, b in zip(list1, list2))
end = time.time()
print("Python dot time:", end - start)
```

```python
# Numpy dot product
start = time.time()
np_dot = np.dot(arr1, arr2)
end = time.time()
print("Numpy dot time:", end - start)
```

```python
# Numpy matrix multiplication (fast)
mat1 = np.random.random((1000, 1000))
mat2 = np.random.random((1000, 1000))

start = time.time()
result_np = mat1 @ mat2
end = time.time()

print("Numpy matrix multiplication time:", end - start)
```
